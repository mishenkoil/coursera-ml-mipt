```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
%matplotlib inline
```

# Проверка ЦПТ на примере распределения арксинуса  
Функция распределения (CDF):  
$F(x) = \frac{2}{\pi}\arcsin\left(\sqrt x\right)=\frac{\arcsin(2x-1)}{\pi}+\frac{1}{2}$  
  
Функция плотности (PDF):  
$f(x) = \frac{1}{\pi\sqrt{x(1-x)}}$

```python
# график плотности нашей случайной величины
x = np.linspace(0.01, 0.99, 100)
pdf = scipy.stats.arcsine().pdf(x)
plt.plot(x, pdf, label='theoretical PDF', color='red')
# гистограмма выборки нашего распределения
sample_cdf = scipy.stats.arcsine().rvs(1000)
plt.hist(sample_cdf, bins=70, label='CDF Histogram', color='pink', density=True)
plt.ylabel('fraction of CDF samples / $F(x)$')
plt.xlabel('$x$')
plt.legend(loc='upper center')
```

Для функции распределения арксинуса: $E(X)=0.5, D(X)=0.125$

**(1)** График плотности соответствующего нормального распределения $(\mu=0.5, \sigma=\sqrt{\frac{0.125}{3}})$:  
**(2)** Гистограмма распределения выборочных средних, выборки объема 3:

```python
# (1):
norm_rv = scipy.stats.norm(0.5, np.sqrt(0.125/3))
norm_pdf = norm_rv.pdf(x)
plt.plot(x, norm_pdf, label='Noraml PDF')
# (2):
sample_mean = []
for i in range(1000):
    sample_mean.append(sum(scipy.stats.arcsine().rvs(3))/3)
plt.hist(sample_mean,bins=40, label='sample mean histogram', color='pink', density=True)
plt.ylabel('sample_mean / normal $F(X)$')
plt.xlabel('$x$')
plt.legend(loc='upper left')
```
